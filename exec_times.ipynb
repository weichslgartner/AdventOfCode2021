```python
from typing import List
import matplotlib.pyplot as plt 
import numpy as np
plt.rcParams.update({'font.size': 16})
plt.style.use('seaborn-pastel')
```

```python
pypy = open("exec_times/pypy.txt",'r').read()
py310 = open("exec_times/py10.txt",'r').read()
py39 = open("exec_times/py39.txt",'r').read()
py36 = open("exec_times/py36.txt",'r').read()
```

```python
def to_list(build_output : str) -> List[str]:
    perf = []
    for line in build_output.split('\n'):
        if line.startswith('real'):
            perf.append( float(line.split()[1][2:-1].replace(',','.')))
    return perf
```

```python
pypy = to_list(pypy)
py310 = to_list(py310)
py39 = to_list(py39)
py36 = to_list(py36)
```

```python
days = np.arange(1,len(pypy)+1)
```

```python
width = 0.15
fig, ax = plt.subplots()
fig.set_size_inches(17, 9)
ax.set_title("Execution Time of Advent of Code 2021 Python Solutions (Both Parts)")
rects1 = ax.bar(days - width*1.5, pypy, width, label='PyPy 7.3')
rects2 = ax.bar(days - width/2, py310, width, label='Python 3.10')
rects3 = ax.bar(days + width/2, py39, width, label='Python 3.9')
rects4 = ax.bar(days + width*1.5, py36, width, label='Python 3.6')
ax.set_xticks(days)
ax.set_ylabel('Execution Time [s]')
ax.set_xlabel('Day')
ax.legend()
```

```python
plt.plot(np.cumsum(pypy), label="pypy")
plt.plot(np.cumsum(py310), label="py 3.10")
plt.plot(np.cumsum(py39), label="py 3.9")
plt.plot(np.cumsum(py36), label="py 3.6")
plt.legend(loc="upper left")
plt.ylabel('Execution Time [s]')
plt.xlabel('Day')
```
